## a. Phân phối của $\hat{\theta} = X_{\text{max}}$:

- Giả sử $X_1, X_2, \dots, X_n$ độc lập và phân phối đều  $ U(0, \theta)$.

- Xác suất $X_{\text{max}} \leq x$ là tích của xác suất $X_i \leq x$ cho từng $X_i$:

$$
P(X_{\text{max}} \leq x) = P(X_1 \leq x) \cdot P(X_2 \leq x) \cdot \dots \cdot P(X_n \leq x) = \left(\frac{x}{\theta}\right)^n, \quad 0 \leq x \leq \theta.
$$

- Suy ra hàm mật độ xác suất (PDF) của $\hat{\theta}$ là:

$$
f_{\hat{\theta}}(x) = \frac{d}{dx} \left( \frac{x}{\theta} \right)^n = n \frac{x^{n-1}}{\theta^n}, \quad 0 \leq x \leq \theta.
$$


```r
# Thiết lập thông số
n <- 50
theta <- 1
B <- 1000  # Số lần bootstrap
```

```r
# Tạo dữ liệu Uniform(0, theta)
set.seed(42)
data <- runif(n, 0, theta)
```

```r
# Xem data
data
```

```r
# f của theta_hat
true_density <- function(x, n, theta) {
  if (x >= 0 && x <= theta) {
    return(n * (x^(n-1)) / (theta^n))
  } else {
    return(0)
  }
}
```

```r
# Tạo giá trị PDF thực
x_vals <- seq(0, 1, length.out = 100) # Tạo dữ liệu 
# Hàm mật độ xác suất
pdf_vals <- sapply(x_vals, true_density, n = n, theta = theta)

x_vals # Dãy x_vals sẽ chứa 100 giá trị nằm trong khoảng từ 0 đến 1, với các bước nhảy đều nhau.
pdf_vals # Giá trị mật độ xác suất cho toàn bộ các giá trị trong x_vals
```

```r
# Tính giá trị MLE
theta_hat <- max(data)
theta_hat
```

```r
# Bootstrap tham số
param_bootstrap <- replicate(B, {
  # Có giả định một phân phối xác suất: trường hợp này là Uni
  # Sử dụng tham số ước lượng từ dữ liệu gốc (MLE) để sinh lại nhiều tập dữ liệu
  # bootstrap bằng cách lấy mẫu ngẫu nhiên từ phân phối giả định
  boot_sample <- runif(n, 0, theta_hat) 
  # Tính toán lại các giá trị thống kê trên mỗi tập bootstrap
  max(boot_sample)
})
```

```r
# Bootstrap phi tham số
nonparam_bootstrap <- replicate(B, {
  # Không giả định bất kỳ phân phối cụ thể nào cho dữ liệu gốc.
  # Tái tạo các tập dữ liệu bootstrap bằng cách lấy mẫu có hoàn lại
  #(resampling) từ dữ liệu gốc
  boot_sample <- sample(data, n, replace = TRUE)
  # Tính toán giá trị thống kê trên các tập bootstrap và xây dựng phân phối
  max(boot_sample)
})
```

```r

# Vẽ biểu đồ
hist(param_bootstrap, probability = TRUE, main = "Parametric Bootstrap",
     xlab = expression(hat(theta)), col = "skyblue", breaks = 20)
lines(x_vals, pdf_vals, col = "red", lwd = 2)

hist(nonparam_bootstrap, probability = TRUE, main = "Nonparametric Bootstrap",
     xlab = expression(hat(theta)), col = "lightgreen", breaks = 20)
lines(x_vals, pdf_vals, col = "red", lwd = 2)
```

### NHẬN XÉT
- Histogram từ parametric bootstrap khớp khá tốt với đường mật độ xác suất thật.
- Điều này cho thấy phương pháp parametric bootstrap có hiệu quả khi phân phối giả định là đúng (ở đây là U(0,θ)).
- Histogram từ nonparametric bootstrap không khớp hoàn toàn với đường mật độ xác suất thật.
- Điều này là do nonparametric bootstrap chỉ sử dụng dữ liệu gốc để lấy mẫu, không tận dụng thông tin về phân phối thật.

## b. Chứng minh:

- $P(\hat{\theta}^* = \hat{\theta}) = 0$ đối với bootstrap tham số.
- $P(\hat{\theta}^* = \hat{\theta}) \approx 0.632$ đối với bootstrap không tham số.

### 1. Bootstrap tham số:
- Với bootstrap tham số, ta sinh lại các dữ liệu bootstrap từ $U(0, \hat{\theta})$, nên $X_{\text{max}}$ trong mẫu bootstrap thường lớn hơn hoặc nhỏ hơn giá trị $\hat{\theta}$ ban đầu.
- Do đó, xác suất $P(\hat{\theta}^* = \hat{\theta}) = 0$.

### 2. Bootstrap phi tham số:
- Xác suất để tất cả các giá trị bootstrap đều nhỏ hơn $\hat{\theta}$ là:
  $$
  P(\hat{\theta}^* = \hat{\theta}) = 1 - \left( 1 - \frac{1}{n} \right)^n.
  $$
- Khi $n \to \infty$, giới hạn của biểu thức này tiến về:
  $$
  \lim_{n \to \infty} \left( 1 - \left( 1 - \frac{1}{n} \right)^n \right) = 1 - e^{-1} \approx 0.632.
  $$


```r
# Hàm tính P(theta_hat* = theta_hat) cho bootstrap không tham số
prob_nonparam <- function(n) {
  return(1 - (1 - 1/n)^n)
}

# Tính xác suất P(theta_hat* = theta_hat) khi n lớn
n_vals <- c(10, 50, 100, 1000)
probs <- sapply(n_vals, prob_nonparam)
probs

# Giới hạn khi n -> Inf
limit <- 1 - exp(-1)
limit
```
