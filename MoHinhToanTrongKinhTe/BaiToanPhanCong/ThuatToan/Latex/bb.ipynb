```python
import numpy as np

# Bước 1: Khởi tạo ma trận chi phí
cost_matrix = np.array([
    [8, 2, 5],
    [3, 2, 7],
    [4, 1, 6]
])

# Bước 2: Tính cận dưới ban đầu bằng cách giảm hàng
def reduce_rows(matrix):
    row_min = np.min(matrix, axis=1)
    reduced_matrix = matrix - row_min[:, np.newaxis]
    return np.sum(row_min), reduced_matrix

# Bước 3: Tính cận dưới mới sau khi loại bỏ hàng và cột
def reduce_columns(matrix):
    col_min = np.min(matrix, axis=0)
    reduced_matrix = matrix - col_min
    return np.sum(col_min), reduced_matrix

# Bước 4: Tính cận dưới ban đầu bằng cách giảm hàng và cột
def calculate_initial_lower_bound(matrix):
    row_reduction_sum, reduced_matrix = reduce_rows(matrix)
    col_reduction_sum, final_reduced_matrix = reduce_columns(reduced_matrix)
    total_reduction_sum = row_reduction_sum + col_reduction_sum
    return total_reduction_sum, final_reduced_matrix

# Bước 5: Gán công việc theo cách Nhánh và Cận

# Bắt đầu tính toán
def branch_and_bound_assignment(cost_matrix):
    total_cost = 0
    
    # Bước 1: Giảm hàng và cột để tìm cận dưới ban đầu
    initial_lower_bound, reduced_matrix = calculate_initial_lower_bound(cost_matrix)
    print(f"Cận dưới ban đầu: {initial_lower_bound}")
    total_cost += initial_lower_bound
    
    # Bước 2: Gán Người 3 cho Công việc 2 (chi phí = 0)
    total_cost += 0  # Gán chi phí 0
    print("Gán Người 3 cho Công việc 2")
    
    # Loại bỏ hàng 3 và cột 2
    reduced_matrix_step2 = np.delete(np.delete(reduced_matrix, 2, axis=0), 1, axis=1)
    print(f"Ma trận sau khi loại bỏ hàng 3 và cột 2:\n{reduced_matrix_step2}")
    
    # Bước 3: Giảm lại hàng và cột trong ma trận còn lại
    row_reduction_sum, reduced_matrix_step3 = reduce_rows(reduced_matrix_step2)
    col_reduction_sum, reduced_matrix_step3 = reduce_columns(reduced_matrix_step3)
    lower_bound_step3 = row_reduction_sum + col_reduction_sum
    total_cost += lower_bound_step3
    print(f"Cận dưới mới sau Bước 3: {lower_bound_step3}")
    
    # Bước 4: Gán Người 2 cho Công việc 1 (chi phí = 0)
    total_cost += 0  # Gán chi phí 0
    print("Gán Người 2 cho Công việc 1")
    
    # Ma trận còn lại sau khi gán (chỉ còn một ô)
    reduced_matrix_step4 = np.delete(np.delete(reduced_matrix_step3, 1, axis=0), 0, axis=1)
    print(f"Ma trận sau khi loại bỏ hàng 2 và cột 1:\n{reduced_matrix_step4}")
    
    # Bước 5: Gán Người 1 cho Công việc 3 (chi phí = 0)
    total_cost += 0  # Gán chi phí 0
    print("Gán Người 1 cho Công việc 3")
    
    # Tổng chi phí của lời giải
    print(f"Tổng chi phí tối ưu: {total_cost}")

# Chạy thuật toán phân công
branch_and_bound_assignment(cost_matrix)
```
