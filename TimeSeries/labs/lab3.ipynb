### DETRENDING

```r
lm_model = lm(gtemp_both ~ time(gtemp_both))
lm_model
```

```r
b0 = lm_model$coefficients[1]
b1 = lm_model$coefficients[2]
```

```r
y_trend = gtemp_both - b0 - b1*(time(gtemp_both))
y_trend
```

## DIFFERENCING

```r
diff_gtemp_both = diff(gtemp_both, lag= 1)
diff_gtemp_both
```

```r
# Set up the plotting area to have 1 row and 2 columns
par(mfrow=c(3,1))

plot(gtemp_both, type = "l", col = "#8aba8f", lwd = 2, ylab = "Values", xlab = "Time", main = "gtemp_both")

plot(y_trend, type = "l", col = "#871c82", lwd = 2, ylab = "Values", xlab = "Time", main = "Detrending")

plot(diff_gtemp_both, type = "l", col = "#b93434", lwd = 2, ylab = "Values", xlab = "Time", main = "Defferencing")
```

```r
mean(diff_gtemp_both)
```

- Khoảng 100 năm thì nhiệt độ trung bình trên trái đất tăng thêm ~1%

```r
acf(diff_gtemp_both)
```

#### 1. White Noise (Nhiễu Trắng)
Định nghĩa: Một chuỗi thời gian được gọi là nhiễu trắng nếu nó bao gồm một chuỗi các biến ngẫu nhiên không có tương quan, mỗi biến có trung bình bằng 0 và phương sai không đổi.
Đặc điểm:
- Trung bình: Bằng 0
- Phương sai: Không đổi
- Tự tương quan: Bằng 0 ở mọi độ trễ (không có tương quan giữa các giá trị ở các thời điểm khác nhau)
- Tính dừng: Hoàn toàn dừng (các đặc tính thống kê không thay đổi theo thời gian)
#### 2. Moving Average (MA) - Trung Bình Trượt
Định nghĩa: Một quá trình MA là một mô hình chuỗi thời gian trong đó giá trị hiện tại phụ thuộc tuyến tính vào các sai số (hoặc nhiễu) trước đó. Mô hình MA(q) biểu thị giá trị hiện tại là sự kết hợp tuyến tính của q sai số trước đó.
Đặc điểm:
- Sự phụ thuộc: Chuỗi phụ thuộc vào một số lượng cố định các sai số trước đó.
- Tự tương quan: Tự tương quan khác 0 lên đến độ trễ q. - Tự tương quan giảm xuống 0 sau độ trễ q.
- Tính dừng: Luôn dừng, vì trung bình và phương sai không thay đổi theo thời gian.
- Trung bình: Không đổi, phụ thuộc vào tham số của mô hình.
- Bộ nhớ: Có bộ nhớ giới hạn về các sai số trước đó; chỉ có q sai số gần nhất ảnh hưởng đến giá trị hiện tại.
#### 3. Autoregressive (AR) - Tự Hồi Quy
Định nghĩa: Một quá trình AR là một mô hình chuỗi thời gian trong đó giá trị hiện tại là sự kết hợp tuyến tính của các giá trị trước đó của chính nó và một sai số ngẫu nhiên. Mô hình AR(p) biểu thị giá trị hiện tại là hàm của p giá trị trước đó.
Đặc điểm:
- Sự phụ thuộc: Giá trị hiện tại phụ thuộc vào một số lượng cố định các giá trị trước đó.
- Tự tương quan: Tự tương quan giảm dần theo thời gian (không giống MA, không giảm xuống 0 sau một số độ trễ nhất định).
- Tính dừng: Mô hình AR có thể dừng hoặc không dừng tùy thuộc vào tham số; điều kiện dừng là tổng trọng số phải nhỏ hơn 1.
- Trung bình: Không đổi (nếu mô hình dừng).
#### 4. Random Walk with a Drift - Đi Bộ Ngẫu Nhiên với Độ Trôi
Định nghĩa: Là một mô hình trong đó giá trị hiện tại là tổng của giá trị trước đó cộng với một sai số ngẫu nhiên và một hằng số (độ trôi).
Đặc điểm:
- Trung bình: Thay đổi theo thời gian, tùy thuộc vào độ trôi.
- Tự tương quan: Tự tương quan rất cao, đặc biệt ở các độ trễ ngắn.
- Tính dừng: Không dừng, vì phương sai tăng theo thời gian.
- Xu hướng: Có thể có xu hướng (nếu có độ trôi).
#### 5. Signal in Noise - Tín Hiệu trong Nhiễu
Định nghĩa: Là một mô hình trong đó chuỗi thời gian bao gồm một phần tín hiệu (có thể dự đoán) và một phần nhiễu (không dự đoán được).
Đặc điểm:
- Tín hiệu: Có thể là một xu hướng hoặc một mẫu có thể dự đoán được.
- Nhiễu: Thường được mô hình hóa như nhiễu trắng.
- Mục tiêu: Tách tín hiệu ra khỏi nhiễu để phân tích.
- Tính dừng: Phần nhiễu thường dừng, nhưng phần tín hiệu có thể không dừng (nếu có xu hướng).

- Đây không phải là một chuỗi thời gian dừng bởi vì:
#### Đặc điểm của Chuỗi Thời Gian Dừng
- Trung bình không đổi: Trung bình của chuỗi thời gian là hằng số và không thay đổi theo thời gian.
- Phương sai không đổi: Phương sai của chuỗi thời gian là hằng số và không phụ thuộc vào thời gian. Điều này có nghĩa là sự biến động của chuỗi thời gian không thay đổi theo thời gian.
- Tự tương quan không đổi: Tự tương quan giữa các giá trị trong chuỗi thời gian chỉ phụ thuộc vào khoảng cách giữa các thời điểm (độ trễ), chứ không phụ thuộc vào thời điểm cụ thể. Điều này có nghĩa là mô hình tương quan giữa các giá trị không thay đổi theo thời gian.

- Hai chuỗi này có xu hướng tăng do đó nó không phải là quá trình dừng
