# ÁP DỤNG DETRENDING

```r
lm_model = lm(chicken ~ time(chicken))
lm_model
```

```r
b0 = lm_model$coefficients[1]
b1 = lm_model$coefficients[2]
```

```r
y_trend = chicken - b0 - b1*(time((chicken)))
y_trend
```

# ÁP DỤNG PHƯƠNG PHÁP DIFFERENCING

```r
diff_checken = diff(chicken, lag= 1)
diff_checken
```

```r
# Set up the plotting area to have 1 row and 2 columns
par(mfrow=c(3,1))

plot(chicken, type = "l", col = "#8aba8f", lwd = 2, ylab = "Values", xlab = "Time", main = "Chicken")

plot(y_trend, type = "l", col = "#871c82", lwd = 2, ylab = "Values", xlab = "Time", main = "Detrending")

plot(diff_checken, type = "l", col = "#b93434", lwd = 2, ylab = "Values", xlab = "Time", main = "Defferencing")
```
