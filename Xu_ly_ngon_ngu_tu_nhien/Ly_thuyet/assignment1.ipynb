```r
# Số lượng bước thời gian
n_steps <- 10

# Tạo một chuỗi ngẫu nhiên chỉ có hai giá trị 0 và 1
random_walk <- sample(c(0, 1), n_steps, replace = TRUE)

# Vẽ biểu đồ của quá trình ngẫu nhiên
plot(random_walk, type = "l", main = "Quá trình ngẫu nhiên (0,1)", 
xlab = "Bước thời gian", ylab = "Giá trị")
```
