- Biểu đồ Residuals vs Fitted: Mục đích: Kiểm tra tính tuyến tính và đồng nhất của phương sai (homoscedasticity) của các phần dư.
Giải thích: Biểu đồ này hiển thị các phần dư trên trục y và các giá trị ước tính trên trục x. Lý tưởng nhất là các phần dư phải phân tán ngẫu nhiên quanh đường ngang (y=0) mà không có mẫu nào cụ thể. Nếu có mẫu hiện rõ (ví dụ: hình phễu), điều này chỉ ra các vấn đề như phi tuyến tính hoặc phương sai không đồng nhất.

- Biểu đồ Q-Q (Quantile-Quantile Plot):- 
Mục đích: Đánh giá tính bình thường của các phần dư.
Giải thích: Biểu đồ này so sánh phân phối của các phần dư với phân phối chuẩn lý thuyết. Nếu các phần dư được phân phối bình thường, các điểm sẽ nằm trên đường chéo. Các lệch lạc so với đường này chỉ ra sự sai lệch khỏi tính bình thường.

- Biểu đồ Scale-Location (hoặc Spread-Location): Mục đích: Kiểm tra tính đồng nhất của phương sai (homoscedasticity) của các phần dư.
Giải thích: Biểu đồ này hiển thị căn bậc hai của các phần dư chuẩn hóa trên trục y và các giá trị ước tính trên trục x. Lý tưởng nhất là các điểm phải phân tán ngẫu nhiên mà không có mẫu rõ ràng. Một mẫu rõ ràng (ví dụ: xu hướng hoặc hình phễu) gợi ý phương sai không đồng nhất.

- Biểu đồ Scale-Location (hoặc Spread-Location): Mục đích: Xác định các điểm dữ liệu có ảnh hưởng lớn.
Giải thích: Biểu đồ này hiển thị các phần dư chuẩn hóa trên trục y và đòn bẩy trên trục x. Các điểm có đòn bẩy cao và phần dư lớn là những điểm có ảnh hưởng tiềm tàng. Đường Cook's distance giúp xác định các điểm có ảnh hưởng này. Các điểm nằm ngoài các đường này (đặc biệt là những điểm có đòn bẩy cao và phần dư cao) có thể có ảnh hưởng đáng kể đến mô hình.
